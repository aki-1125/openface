# 1. OpenFaceの生データ前処理

```python
#データの表示
data_a = pd.read_csv("OpenFace/20201111_session1_A_gvo.csv")
data_b = pd.read_csv("OpenFace/20201111_session1_B_gvo.csv")
data_c = pd.read_csv("OpenFace/20201111_session1_C_gvo.csv")
data_d = pd.read_csv("OpenFace/20201111_session1_D_gvo.csv")
```

```python
#各列に空白が入ってしまっているので消去
data_a.columns = data_a.columns.str.strip()
data_b.columns = data_b.columns.str.strip()
data_c.columns = data_c.columns.str.strip()
data_d.columns = data_d.columns.str.strip()
```

```python
#時系列データを扱うためのライブラリの読み込み
import datetime as dt
```

```python
#変数a,b,c,dにtimestampの列をhh:mm:ss.f形式にしたものを入れていく
a = []
for i in range(0, len(data_a)):
    time_a = str(dt.timedelta(seconds=data_a["timestamp"][i]))
    a.append(time_a)

b = []
for i in range(0, len(data_b)):
    time_b = str(dt.timedelta(seconds=data_b["timestamp"][i]))
    b.append(time_b)

c = []
for i in range(0, len(data_c)):
    time_c = str(dt.timedelta(seconds=data_c["timestamp"][i]))
    c.append(time_c)

d = []
for i in range(0, len(data_d)):
    time_d = str(dt.timedelta(seconds=data_d["timestamp"][i]))
    d.append(time_d)
```

```python
#元のデータのtimestampの列に変数a,b,c,dを代入
data_a.timestamp = a
data_b.timestamp = b
data_c.timestamp = c
data_d.timestamp = d
```

```python
#編集データ表示
data_d
```

```python
#timestampの列をdatetime型にする
#時間のフォーマットが一部違うため，引数infer_datetime_format=Trueを指定している
data_a['timestamp'] = pd.to_datetime(data_a['timestamp'], infer_datetime_format=True)
data_b['timestamp'] = pd.to_datetime(data_b['timestamp'], infer_datetime_format=True)
data_c['timestamp'] = pd.to_datetime(data_c['timestamp'], infer_datetime_format=True)
data_d['timestamp'] = pd.to_datetime(data_d['timestamp'], infer_datetime_format=True)
```

```python
#編集データ表示
data_a
```

```python
#時刻パラメータの設定
d1 = dt.datetime(year=2020, month=11, day=11) #実験実施日
d2 = dt.datetime.today() #今日
days_order = d1 - d2 #日数計算(pd.to_datetimeのデフォルトで日程の部分が今日になっているので実験実施日-今日を行っている)
days_order = days_order.days + 1 #経過日数を表示するので，1を足す必要がある(例えば8月15日から8月20日の日数の数え方は15日を入れるかで変わる)
hours_order = 12 #時を指定
miniute_order = 40 #分を指定
seconds_order = 3 #秒を指定
```

```python
#設定した時刻パラメータを用いてtimedelta関数で時刻計算を行う
data_a['timestamp'] = data_a['timestamp'] + dt.timedelta(days=days_order, hours=hours_order, minutes=miniute_order, seconds=seconds_order)
data_b['timestamp'] = data_b['timestamp'] + dt.timedelta(days=days_order, hours=hours_order, minutes=miniute_order, seconds=seconds_order)
data_c['timestamp'] = data_c['timestamp'] + dt.timedelta(days=days_order, hours=hours_order, minutes=miniute_order, seconds=seconds_order)
data_d['timestamp'] = data_d['timestamp'] + dt.timedelta(days=days_order, hours=hours_order, minutes=miniute_order, seconds=seconds_order)
```

```python
#編集データ表示
data_a
```

```python
#timestampの列をインデックス化
data_a.set_index("timestamp", inplace=True)
data_b.set_index("timestamp", inplace=True)
data_c.set_index("timestamp", inplace=True)
data_d.set_index("timestamp", inplace=True)
```
